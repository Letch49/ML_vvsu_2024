#Основы статистики и первичная визуализация

##Введение в описательную статистику

###Теория

Описательная статистика - раздел статистики, который предоставляет инструменты и методы для описания и анализа данных. Она играет важную роль в машинном обучении, статистическом анализе и научных исследованиях, помогая понять характеристики и распределение данных без необходимости глубокого анализа. Описательная статистика помогает выявить основные характеристики, выбросы, идентифицировать закономерности и подготовить данные для более сложных анализов.

Основные показатели: среднее, медиана, мода, минимум, максимум:


* Среднее (среднее арифметическое): Это сумма всех значений, деленная на количество значений. Среднее позволяет оценить "типичное" значение в данных. Однако оно может быть подвержено влиянию выбросов.
* Медиана: Медиана - это значение, которое разделяет набор данных на две равные половины. Она устойчива к выбросам и дает представление о центральной "средине" данных.
* Мода: Мода - это значение, которое встречается наиболее часто в наборе данных. Мода полезна для идентификации наиболее частых значений.
* Минимум и максимум: Минимум - это наименьшее значение в данных, а максимум - наибольшее. Они предоставляют информацию о диапазоне значений в данных.

Квантили и интерквартильный размах:

Квантили: Квантили разделяют данные на части определенных долей. Например, медиана является вторым квантилем, разделяющим данные пополам. Первый квантиль (25-й процентиль) отделяет наименьшие 25% данных, третий квантиль (75-й процентиль) отделяет наименьшие 75% данных.

Интерквартильный размах: Интерквартильный размах - это разница между третьим и первым квантилем. Он предоставляет информацию о разбросе средних значений 50% данных в середине распределения.

###Пример

```python
import pandas as pd
df = pd.read_csv("/content/sample_data/california_housing_test.csv")
df
```

```python
# Вычисление квантилей
q1 = df['latitude'].quantile(0.25)
q2 = df['latitude'].quantile(0.50)  # Медиана
q3 = df['latitude'].quantile(0.75)

# Вычисление интерквартильного размаха
iqr = q3 - q1

# Вычисление медианы, среднего, максимума и минимума
median = df['latitude'].median()
mean = df['latitude'].mean()
minimum = df['latitude'].min()
maximum = df['latitude'].max()

print("Первый квантиль (25%):", q1)
print("Второй квантиль (медиана) (50%):", q2)
print("Третий квантиль (75%):", q3)
print("Интерквартильный размах:", iqr)
print("Медиана (второй квантиль):", median)
print("Среднее:", mean)
print("Минимум:", minimum)
print("Максимум:", maximum)
```

##Основы визуализации данных



###Теория


Визуализация данных - процесс представления информации в графической форме, что позволяет обнаруживать паттерны, тренды, аномалии и взаимосвязи между данными, которые могут быть трудно обнаружить при анализе сырых числовых значений. Визуализация делает данные более понятными, помогает обосновать выводы и принимать информированные решения.

Типы графиков: гистограммы, Boxplot, диаграмма рассеивания:

Гистограммы:
Гистограмма - график, который показывает распределение частот или относительных частот значений в числовом наборе данных. Ось X представляет значения, а ось Y - частоту или относительную частоту. Гистограммы полезны для понимания формы распределения данных, выявления пиков и выбросов.

Boxplot (ящик с усами):
Boxplot - это график, который позволяет визуализировать основные статистические показатели данных, такие как медиана, квартили и выбросы. Ящик представляет интерквартильный размах, а усы - данные вне этого диапазона. Boxplot помогает выявить асимметрию данных, наличие выбросов и распределение.

Диаграмма рассеивания (scatter plot):
Диаграмма рассеивания показывает точки данных на плоскости, где каждая точка представляет пару значений двух переменных. Это позволяет наблюдать взаимосвязь между переменными, идентифицировать корреляции (положительные, отрицательные) и выявлять выбросы.

Знание и использование различных типов графиков позволяют более полно исследовать и анализировать данные, делая важные выводы и принимая обоснованные решения.

*p.s. Выбросы - это значения в наборе данных, которые существенно отличаются от остальных значений и могут искажать общую картину распределения. В контексте boxplot, выбросы представляют собой значения, которые лежат за пределами "усов" графика и могут указывать на наличие аномальных или нетипичных наблюдений.*

###Используемые библиотеки

Использование библиотек визуализации, таких как Matplotlib и Seaborn:

Для создания графиков в Python широко используются библиотеки визуализации данных, такие как Matplotlib и Seaborn.

Matplotlib: Это мощная библиотека для создания разнообразных графиков, от простых до сложных. Она предоставляет полный контроль над внешним видом графиков и диаграмм. Matplotlib позволяет создавать линейные графики, столбчатые диаграммы, точечные графики и др.

Seaborn: Это более высокоуровневая библиотека, основанная на Matplotlib. Она предоставляет более удобный интерфейс для создания стильных и информативных графиков. Seaborn часто используется для создания графиков плотности, гистограмм, violin plots и диаграмм рассеивания с различными стилями.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Данные о возрасте и росте
data = {
    'age': [22, 28, 30, 35, 40, 42, 45, 48, 50, 55],
    'height': [165, 170, 160, 175, 180, 178, 172, 168, 162, 170]
}

# Пример гистограммы для возраста
plt.figure(figsize=(4, 3))
sns.histplot(data['age'], bins=6, edgecolor='black')
plt.title('Гистограмма возраста с Seaborn')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.show()

# Пример Boxplot для возраста
plt.figure(figsize=(4, 3))
sns.boxplot(data=data['age'], color='lightgreen')
plt.title('Boxplot возраста с Seaborn')
plt.xlabel('Возраст')
plt.show()

# Пример диаграммы рассеивания для возраста и роста
data_df = pd.DataFrame(data)
plt.figure(figsize=(4, 3))
sns.scatterplot(x='age', y='height', data=data_df, color='purple')
plt.title('Диаграмма рассеивания для возраста и роста с Seaborn')
plt.xlabel('Возраст')
plt.ylabel('Рост')
plt.show()
```

##Корреляция в статистике

###Теория

Корреляция - статистическая мера, которая измеряет степень взаимосвязи между двумя или более переменными. Если изменения в одной переменной соотносятся с определенными изменениями в другой переменной, то между ними существует корреляция.

Типы корреляции:

Положительная корреляция: Если увеличение значений одной переменной сопровождается увеличением значений другой переменной, то между ними существует положительная корреляция. Коэффициент корреляции в этом случае будет положительным.

Отрицательная корреляция: Если увеличение значений одной переменной связано с уменьшением значений другой переменной, то между ними существует отрицательная корреляция. Коэффициент корреляции в этом случае будет отрицательным.

Отсутствие корреляции (нулевая корреляция): Если изменения в одной переменной не связаны с какими-либо систематическими изменениями в другой переменной, то между ними отсутствует корреляция. Коэффициент корреляции близок к нулю.

Коэффициент корреляции:

Коэффициент корреляции измеряет степень и направление взаимосвязи между переменными. Наиболее распространенными методами вычисления коэффициента корреляции являются коэффициент Пирсона и коэффициент Спирмена.

Коэффициент Пирсона: Измеряет линейную зависимость между переменными. Его значения находятся в диапазоне от -1 до 1, где -1 означает полную отрицательную корреляцию, 1 - положительную корреляцию, а 0 - отсутствие корреляции.

Коэффициент Спирмена: Оценивает степень монотонной зависимости между переменными. Он также находится в диапазоне от -1 до 1, где -1 и 1 указывают на полную монотонную зависимость, а 0 - на отсутствие зависимости.

При вычислении коэффициента корреляции также важно оценивать его значимость. Важно понимать, что даже если коэффициент корреляции высок, это не гарантирует наличие реальной связи между переменными. Случайность может привести к высокому значению корреляции. Поэтому часто проводят статистические тесты для оценки значимости корреляции.

###Монотонная зависимость


Монотонная зависимость между двумя переменными означает, что при изменении значений одной переменной, значения другой переменной либо возрастают, либо убывают. Важно отметить, что монотонная зависимость не обязательно является линейной. Это означает, что изменения переменных могут происходить в более сложных паттернах, но всегда с однозначным направлением.

###Мультиколлинеарность


Мультиколлинеарность - это статистическое явление, при котором две или более независимых переменных в модели имеют высокую корреляцию между собой. Это может создать проблемы при анализе данных и построении моделей, так как мультиколлинеарность может искажать результаты оценок и усложнять интерпретацию влияния переменных на зависимую переменную.

Причины мультиколлинеарности:

* Линейные зависимости: Если одна независимая переменная может быть линейно представлена как линейная комбинация другой переменной (или нескольких переменных), это может привести к мультиколлинеарности.

* Масштабирование: Разный масштаб переменных может привести к высокой корреляции между ними. Например, если одна переменная измеряется в тысячах, а другая - в долях, они могут быть коррелированы из-за разницы в масштабе.

* Соотношения между переменными: Если переменные связаны каким-либо физическим или теоретическим соотношением, это также может привести к мультиколлинеарности.

Последствия мультиколлинеарности:

Сложности в интерпретации: Высокая корреляция между переменными усложняет определение их независимого влияния на зависимую переменную.

Неустойчивость оценок: Мультиколлинеарность может привести к высокой дисперсии оценок коэффициентов модели, что делает их менее надежными.

Проблемы с обратной матрицей: В случае мультиколлинеарности, обратная матрица, используемая для оценки коэффициентов, может стать неустойчивой или несуществующей.
