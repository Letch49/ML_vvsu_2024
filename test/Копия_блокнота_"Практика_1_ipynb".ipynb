##Форматы данных (теория)

###Формат JSON (JavaScript Object Notation):
JSON - это текстовый формат обмена данными, который широко используется для передачи и хранения структурированных данных. Он основан на синтаксисе JavaScript, но может быть использован в разных языках программирования. Формат JSON поддерживает объекты, массивы, строки, числа, логические значения и null.
В Python ближе всего к JSON - словари. В примере name, age, city - ключи. John, 30, New York - значения

````python
example_json_object = {
  "name": "John",
  "age": 30,
  "city": "New York"
}
````

###Формат CSV (Comma-Separated Values):
CSV - это текстовый формат, используемый для представления таблиц данных. В CSV-файлах значения разделяются запятыми или другими символами (такими как точка с запятой). Он часто используется для обмена табличными данными между программами. В примере name, age, city - названия колонок, остальное - значения в колонках

````python
name,age,city
John,30,New York
Alice,25,San Francisco
````

###Формат XLSX (Excel Workbook):
XLSX - это формат файла, используемый в Microsoft Excel для хранения табличных данных. Файлы XLSX могут содержать несколько листов, каждый из которых может содержать таблицы, графику и другие элементы. С ним все знакомы.

###SQL (Structured Query Language):
SQL - это язык программирования, используемый для работы с реляционными базами данных. Он позволяет создавать, изменять, удалять и извлекать данные из баз данных. SQL оперирует различными операторами, такими как SELECT (извлечение данных), INSERT (вставка данных), UPDATE (обновление данных) и DELETE (удаление данных). В реляционных базах данные представлены в виде таблиц, аналогичных csv.

###Общие принципы работы с данными в форматах:

JSON, CSV и XLSX - это текстовые форматы данных, которые можно легко читать и записывать в большинстве языков.
Формат JSON хорошо подходит для хранения и передачи структурированных данных, таких как конфигурации, API-ответы и др.
Формат CSV удобен для обмена таблицами данных, но может не поддерживать сложные структуры данных.
Формат XLSX обычно используется для хранения и анализа данных в Excel, но также может быть читаемым в программе с поддержкой.

Запросы SQL позволяют извлекать информацию, модифицировать данные и управлять структурой базы данных.

##Работа с различными форматами
Чаще всего в мире данных используется библиотека pandas, которая является универсальной и работает со всеми ранее представленными типами данных
Библиотека pandas - это мощный инструмент в языке программирования Python, предназначенный для анализа и обработки данных. Она предоставляет структуры данных и функции для эффективной манипуляции с табличными данными, анализа временных рядов, загрузки и сохранения данных в различных форматах. В основе pandas лежат две основные структуры данных: Series и DataFrame.

Series: Это одномерная индексированная структура данных, представляющая собой массив значений. Каждый элемент Series сопоставлен с определенным индексом, что позволяет быстро и легко обращаться к данным.

DataFrame: Это двумерная структура данных, представляющая собой таблицу с рядами и столбцами. DataFrame позволяет хранить и обрабатывать данные в табличной форме, а также обеспечивает гибкую индексацию и множество методов для работы с данными.

#Признаки
Для работы с объектом модель должна опираться на какие-то его свойства, например, доход человека, цвет левого верхнего пикселя на изображении или частоту встречаемости слова «интеграл» в тексте. Эти свойства обычно называются признаками, а совокупность свойств, которые мы выделили у объекта – его признаковым описанием.

Численные – например, рост или доход. Иногда отдельно выделяют вещественные и целочисленные признаки.

Категориальные признаки (номинальные) принимают значения из некоторого дискретного множества. Например, профессия человека или день недели.

Бинарные признаки принимают два значения:
 или «да» и «нет». С ними можно работать и как с численными, и как с категориальными.

В нашем случае

Series - категория

X и Y - численные
